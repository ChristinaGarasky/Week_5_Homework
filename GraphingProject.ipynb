```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from pprint import pprint
```

```python
mydata = pd.read_csv("SAheartdisease.csv",)
print(mydata)
```

```python
Age_obese = plt.scatter(mydata["age"], mydata["obesity"])
plt.title("Age vs. Obesity")
plt.ylabel("Obesity")
plt.xlabel("Age")
plt.show(Age_obese)
```

```python

plt.boxplot(mydata['sbp']) 
plt.title("Box plot of Systolic Blood Pressure")
plt.ylabel("Systolic Blood Pressure")
# show plot 
plt.show() 
```

```python
#Most people in this sample do not smoke. 
#The people who do smoke, they smoke less then 10kg a day. 
#Very few are "chain" smokers who smoke about 10kg a day. 
new = plt.hist(mydata['tobacco'], bins='auto')
plt.title("Histogram of Tobacco usage (kg)")
plt.ylabel("Count")
plt.xlabel("Tobacco usage")
plt.show(new)
```

```python
#Scatter plot of tobacco use and obesity. 
#You would need to run a correlation, but it looks like their may be a small positive corrrelation between obesity and tobacco use. 

Tobacco_obese = plt.scatter(mydata["tobacco"], mydata["obesity"])
plt.title("Tobacco Use (kg) vs. Obesity")
plt.ylabel("Obesity")
plt.xlabel("Tobacco")
plt.show(Tobacco_obese)
```
